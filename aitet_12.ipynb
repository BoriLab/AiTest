```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(color_codes=True)
%matplotlib inline
```

```python
df = pd.read_csv("/content/drive/MyDrive/hotel_bookings.csv")

# 데이터 샘플 확인
df.head()
```

```python
# 특정 변수 분포 시각화
plt.rcParams['figure.dpi'] = 300
sns.distplot(df['lead_time'])
```
